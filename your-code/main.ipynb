#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
steps_months_encoded = fitbit['Steps'].groupby(fitbit['Months_encoded']).median()
# MEDIAN STEPS BY WORK_OR_WEEKEND
steps_work_or_weekend = fitbit['Steps'].groupby(fitbit['Work_or_Weekend']).median()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
plt.sca(ax1)
steps_months_encoded.plot(kind = 'line', color = 'b', linewidth=4)
plt.xticks(range(15), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], color='b')
plt.title('Median steps by month encoded', fontsize=20, color='b')
plt.xlabel('Months encoded', fontsize=20, color='b')
plt.ylabel('Steps', fontsize=20, color='b')
plt.grid(True)

plt.sca(ax2)
steps_work_or_weekend.plot(kind = 'bar', alpha = 0.8, color='r')
plt.xticks(range(3), ['Weekdays', 'Weekend'], color='r')
plt.title('Median steps by work or weekend', fontsize=18,color='r')
plt.xlabel('Work or weekend', fontsize=18,color='r')
plt.ylabel('Steps', fontsize=18, color='r')
plt.grid(True)

plt.savefig("figures_ch2.jpg")
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
x = ['Minutes Sedentary', 'Minutes Lightly Active','Minutes Very Active']
y= fitbit['Steps']
for var in x:
    figure = plt.figure
    ax = plt.gca()
    ax.scatter(fitbit[var], y)
    ax.set_xlabel(var)
    ax.set_ylabel('Steps')
    ax.set_title(f"{var} vs. Steps")
    plt.show()
```
